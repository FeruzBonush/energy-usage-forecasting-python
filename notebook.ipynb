file_path = "data/household_power_consumption.txt"


```python
import pandas as pd

df = pd.read_csv(
    "data/household_power_consumption.txt",
    sep=';',
    na_values='?',
    low_memory=False,
    parse_dates={'datetime': ['Date', 'Time']},
    dayfirst=True
)

df = df.dropna(subset=['Global_active_power'])

# Keep only first ~30 days (small & fast)
sample = df[df['datetime'] < df['datetime'].min() + pd.Timedelta(days=30)]

sample.to_csv("data/sample_household_power_consumption.csv", index=False)

print("Sample dataset saved to data/sample_household_power_consumption.csv")
```
